## Importing Necessary Liberaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



## to ignore warnings

import warnings as wg
wg.filterwarnings("ignore")
```

## Importing Data.

```python
data = pd.read_csv("SampleSuperstore.csv")
data.head()
```

```python
data.shape  ## Shape Function is used to know the number of rows and columns in data set
```

```python
data.info() # Info function is used to know the null value in the data set
```

```python
data.describe()
```

```python
## Checking for duplicate values

data.duplicated().sum()
```

```python
# Dropping duplicate values

data.drop_duplicates()
data.head()
```

```python
# Removing the unnecessary column

dataset = data.drop(['Postal Code'], axis=1)
dataset.head()
```

## Visualising the  category dataset through pairplot 

```python
# visualising the dataset through pairplot 

sns.pairplot(dataset, hue= 'Category')

```

##  Now finding the pairwaise correlation between the column and visualising the correlation through heatmap.


```python
# Finding the pairwaise correlation between the column and visualising the correlation through heatmap
import seaborn as sns
dataset.corr()
plt.figure(figsize= (10,12))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

##  Visualising the category

```python
plt.figure(figsize = (10,10))
textprops = {"fontsize":15}
plt.title('Category', fontsize= 20)
plt.pie(dataset['Category'].value_counts(), labels=dataset['Category'].value_counts().index,
        autopct='%1.1f%%',textprops = textprops)
plt.show()
```

## Visualising the Sub-Category

```python
plt.figure(figsize = (10,10))
textprops = {"fontsize":15}
plt.title('Sub-Category', fontsize= 20)
plt.pie(dataset['Sub-Category'].value_counts(), labels=dataset['Sub-Category'].value_counts().index,
        autopct='%1.1f%%',textprops = textprops)
plt.show()
```

## Now Finding States which have highest sales

```python

State_sales = dataset.groupby('State').sum()['Sales'].sort_values(ascending = False).reset_index()
plt.figure(figsize=(15,10))
plt.title('States with highest sales rate', fontsize= 25)
sns.barplot(x = 'State', y = 'Sales', data = State_sales.head(15)) 
plt.xticks(rotation = 90 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('States', fontsize= 20)
plt.ylabel('Sales', fontsize= 20)
plt.show()
```

## Now finding States which have lowest sales

```python

State_sales = dataset.groupby('State').sum()['Sales'].sort_values(ascending = False).reset_index()
plt.figure(figsize=(15,10))
plt.title('States with lowest sales rate', fontsize= 25)
sns.barplot(x = 'State', y = 'Sales', data = State_sales.tail(15)) 
plt.xticks(rotation = 90 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('States', fontsize= 20)
plt.ylabel('Sales', fontsize= 20)
plt.show()
```

## Now finding States which have highest profit.

```python

State_profit = dataset.groupby('State').sum()['Profit'].sort_values(ascending = False).reset_index()
plt.figure(figsize=(15,10))
plt.title('States with highest Profit rate', fontsize= 25)
sns.barplot(x = 'State', y = 'Profit', data = State_profit.head(15)) 
plt.xticks(rotation = 90 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('States', fontsize= 20)
plt.ylabel('Profit', fontsize= 20)
plt.show()
```

## Now finding States which have lowest profit

```python
State_profit = dataset.groupby('State').sum()['Profit'].sort_values(ascending = False).reset_index()
plt.figure(figsize=(15,10))
plt.title('States with lowest Profit rate', fontsize= 25)
sns.barplot(x = 'State', y = 'Profit', data = State_profit.tail(15)) 
plt.xticks(rotation = 90 , fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('States', fontsize= 20)
plt.ylabel('Profit', fontsize= 20)
plt.show()
```

## Now Finding Category wise Sales & Profit

```python
Sales = dataset.groupby('Category').sum()['Sales'].nlargest(n=15) 
Profits = dataset.groupby('Category').sum()['Profit'].nlargest(n=15) 
Sales.plot(kind='bar', figsize=(12,6),fontsize = 15, color = 'orange')
Profits.plot(kind='bar', figsize=(12,6),fontsize = 15 , color='blue')
plt.legend(loc='upper right', fontsize= 15)
plt.xlabel('Category',fontsize =15)
plt.ylabel('Total sales and profit',fontsize = 15)
plt.title(" Category wise Sales & Profit",fontsize = 25)
plt.show()
```

## Now trying to find out profit/loss and sales of each region

```python
dataset.groupby('Region')[['Profit','Sales',]].sum().plot.bar(figsize=(12,8))
plt.xlabel('Region', fontsize= 15)
plt.ylabel('Profit/Loss and sales', fontsize= 15)
plt.legend(loc='upper left', fontsize= 12)

plt.show()
```
